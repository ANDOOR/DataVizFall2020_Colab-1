```
import pandas as pd
```

```
elec_res_2020 = pd.read_csv("2020_US_County_Level_Presidential_Results.csv")
elec_res_past = pd.read_csv("2008_2016_US_County_Level_Presidential_Results.csv")
```

```
elec_res_2020
```

```
elec_res_past[elec_res_past["county"] == "Delta County"]
```

```
elec_res = elec_res_2020.merge(elec_res_past,left_on="county_fips",right_on="fips_code")
```

```
elec_res_agg = elec_res.groupby("state_name", as_index=False).sum()
```

```
elec_res.head()
```

```
elec_res_agg_2020 = elec_res_agg[['state_name','votes_gop',	'votes_dem',	'total_votes']]
```

```
elec_res_agg_past = elec_res_agg[['state_name','total_2008',	'dem_2008',	'gop_2008',	'oth_2008',	'total_2012',	'dem_2012',	'gop_2012',	'oth_2012',	'total_2016',	'dem_2016',	'gop_2016',	'oth_2016']]
```

```
elec_res_agg_2020.to_csv("elec_res_agg_2020.csv",index=False)
```

```
elec_res_agg_past.to_csv("elec_res_agg_past.csv",index=False)
```
